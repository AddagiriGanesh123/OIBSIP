```python
# To consider the data value in descending order so we have mentioned reverse particularly.
df['Selling_Price']
sorted(df['Selling_Price'],reverse=True)#To get Descenging Values
```

```python
#Fetch data values
s=df[~(df['Selling_Price']>=33.0)&(df['Selling_Price']<=35.0)]
s
#Here tilde ~ indicates to print the other values rather than expexcted value,if we remove the tilde value then we get only sinle row valuies in the datset.
#we Successfully removed 2categorical columns
```

```python
df['Fuel_Type'].unique()
```

```python
#Data Encoding-i.e., convert the string datatype to int datatype
df['Fuel_Type']=df['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
df['Fuel_Type'].unique()
```

```python
df['Selling_type'].unique()
```

```python
df['Selling_type']=df['Selling_type'].map({'Dealer':0,'Individual':1})
df['Selling_type'].unique()
```

```python
df['Transmission'].unique()
```

```python
df['Transmission']=df['Transmission'].map({'Manual':0,'Automatic':1})
df['Transmission'].unique()
```

```python
#Store Feature matrix in x and Response in y
x=df.drop(['Car_Name','Selling_Price'],axis=1)
y=df['Selling_Price']
```

```python
x
```

```python
y##To determine about the values related to only Selling_Price column.
```

```python
#Splitting the data into Traning set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=32)
```

```python
#Import the models
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#Fit the model
model.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
random=RandomForestRegressor()
```

```python
random.fit(x_train,y_train)
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor()
```

```python
xgb.fit(x_train,y_train)
```

```python
#Prediction on Test Data
y_pred=model.predict(x_test)
y_pred
```

```python
pred_1=xgb.predict(x_test)
pred_1
```

```python
pred_2=random.predict(x_test)
pred_2
```

```python
#Evaluating the Algorithm-Accuracy Evaluation
from sklearn import metrics
score1=metrics.r2_score(y_test,y_pred)
score2=metrics.r2_score(y_test,pred_1)
score3=metrics.r2_score(y_test,pred_2)
```

```python
print(score1,score2,score3)
```

```python
final_df=pd.DataFrame({'Models':['LR','RF','XGB'],'R2_SCORE':[score1,score2,score3]})
final_df
```

```python
#Save the model
xg=XGBRegressor()
xg_final=xg.fit(x,y)
xg_final
```

```python
import joblib
joblib.dump(xg_final,'car_price_predictor')
```
