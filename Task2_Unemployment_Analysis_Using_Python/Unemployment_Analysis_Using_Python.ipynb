```python
#Rename all the columns
s.columns=["Region","Date","Frequency","Estmiated Unemployment Rate (%)","Estimated Employed","Estimated Participation Rate (%)","Area"]
```

```python
print(s)
```

```python
#Visualisation-correlation between the dataset features.
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(12,10))#12-width,10-height
sns.heatmap(s.corr())
plt.show()
```

```python
#Visualise the data to analyse the unemployment rate fot this we have to consider estimated employment for region particularly
#Rename all the columns
s.columns=["Region","Date","Frequency","Estmiated Unemployment Rate (%)","Estimated Employed","Estimated Participation Rate (%)","Area"]
plt.title("Unemployment in India")
sns.histplot(x="Estmiated Unemployment Rate (%)",hue="Region",data=s)
plt.show()
```

```python
sns.histplot(x="Estimated Participation Rate (%)",hue="Region",data=s)
```

```python
#See the unemployment rate according to different regions of India
plt.figure(figsize=(12,10))
plt.title(" Estimation of Indian Employment")
sns.histplot(x="Estimated Employed",hue="Region",data=s)
plt.show()
```

```python
#create a dashboard to analyse the unemployment rate of each indian state by region
#Here we use sunburst plot
uemp=s[["Area","Region","Estmiated Unemployment Rate (%)"]]
fig=px.sunburst(uemp,path=["Region","Area"],values="Estmiated Unemployment Rate (%)",width=700,height=700,color_continuous_scale="RdY1Gn",title="Employment Rate In India")
fig.show()
```

```python
#Regions -Countplot assured by Regions on the dataframe
sns.countplot(x=df['Region'],palette="bright")
plt.xticks(rotation="vertical")
plt.show()
```

```python
#Boxplot Visualisation of Estmiated Unemployment Rate (%)
plt.figure(figsize=(24,7))
sns.boxplot(x='Region',y='Estmiated Unemployment Rate (%)',data=s)
plt.xticks(rotation='vertical')
plt.title('Unemployment Rate')
plt.show()
```
